```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
file_path = r'C:\Users\mohammad alsarese\OneDrive - Tafila Technical University\carrefour Market cap 21-24(Sheet1) (4).csv'
carrefour_df = pd.read_csv(file_path)
carrefour_df.head()
```

```python
plt.figure(figsize=(15, 8))
sn.lineplot(carrefour_df, x='Month - Year', y='Market_cap', marker='o', label='Market Cap', color='blue')
plt.title('Market Capitalization Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Market Cap (in Billion)', fontsize=14)
plt.grid(alpha=0.3)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
