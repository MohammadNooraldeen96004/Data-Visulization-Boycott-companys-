```python
import pandas as pd
import seaborn as sns
```

```python
cap_df=pd.read_csv(r"C:\Users\DELL\Downloads\AnyConv.com__Market_Cap_Pepsico2.csv")
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
```

```python
# Convert the 'year' column to a datetime format for better plotting
cap_df['year'] = pd.to_datetime(cap_df['year'])

# Plotting the time series data
plt.figure(figsize=(12, 6))
plt.plot(cap_df['year'], cap_df['Market Cap'], marker='o', linestyle='-', color='b', label='Market Cap')

# Adding titles and labels
plt.title('PepsiCo Market Cap Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Market Cap (in billions)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()

# Display the plot
```
